```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Get Pokedex

```python
URL = "https://pokemondb.net/pokedex/all"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
rows = soup.find_all("tr") #,class_="cell-num cell-fixed")
rows = soup.find("table", id="pokedex").find("tbody").find_all("tr")
cols = ['number', 'name', 'type1', 'type2','Total', 'HP', 'Att', 'Def', 'SPAtt', 'SPDef', 'Speed']
pokedex = pd.DataFrame(columns=cols)

for row in rows:
    cells = row.find_all("td")
    number = cells[0].get_text()
    name = cells[1].get_text()
    type1 = cells[2].get_text().split(" ")[0]
    type2 = cells[2].get_text().split(" ")[1]
    stats1 = cells[3].get_text()
    stats2 = cells[4].get_text()
    stats3 = cells[5].get_text()
    stats4 = cells[6].get_text()
    stats5 = cells[7].get_text()
    stats6 = cells[8].get_text()
    stats7 = cells[9].get_text()
    aux = pd.DataFrame([[number, name, type1, type2, stats1, stats2, stats3, stats4, stats5, stats6, stats7]], columns = cols)
    pokedex = pd.concat([pokedex, aux])
```
